```python
data_cleaned = data_cleaned.drop(['House price of unit area'],axis=1)
```

### Checking the multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables = data_cleaned[['House Age','Distance from nearest Metro station (km)','Number of convenience stores']]
```

```python
vif = pd.DataFrame()
```

```python
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
```

```python
vif["Features"] = variables.columns
```

```python
vif
```

# Linear regression Model

```python
targets = data_cleaned['log_price']
inputs = data_cleaned.drop(['log_price'],axis=1)
```

### Train and Test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)

plt.xlabel('Target y_train', size=18)
plt.ylabel('Predictions y_hat', size=18)

plt.show()
```

```python
sns.distplot(y_train - y_hat)

# Include a title
plt.title("Residuals PDF", size=18)
```

```python
reg.score(x_train,y_train)
```

### Finding weight and biases

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary["Weights"]= reg.coef_
reg_summary
```

### Testing the Model

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test)
plt.xlabel('Targets y_test',size=18)
plt.ylabel('Predictions y_hat_test', size=18)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test = y_test.reset_index(drop=True)

# Check the result
y_test.head()
```

```python
# Again, we need the exponential of the test log price
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
pd.options.display.max_rows = 999
# Moreover, to make the dataset clear, we can display the result with only 2 digits after the dot 
pd.set_option('display.float_format', lambda x: '%.2f' % x)
# Finally, we sort by difference in % and manually check the model
df_pf.sort_values(by=['Difference%'])
```
